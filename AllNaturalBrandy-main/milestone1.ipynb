Returns the primary title and whether or not it's an adult film for movies that are adult films and the genre of movie is shared between the hollywood and bollywood datasets. It is ordered by the primary title and limited to 10 results.

```python
%%bigquery
select distinct primaryTitle AS pt, isAdult
from IMDb_staging.title_basics AS hollywood
JOIN kaggle_staging.bollywood AS bollywood
ON hollywood.genres = bollywood.genre
WHERE isAdult = "1"
ORDER BY pt
LIMIT 10
```

04/15/2021
Returns the region for movies where hollywood title begins with 'A'
and the title of movie is shared between the hollywood and bollywood
datasets. It is ordered by the title and limited to 15 results.

```python
%%bigquery
select region AS r
from IMDb_staging.title_akas AS hollywood
JOIN kaggle_staging.bollywood AS bollywood
ON hollywood.title = bollywood.title
WHERE hollywood.title LIKE 'A%'
ORDER BY hollywood.title
LIMIT 15
```

04/15/2021
Returns the primary title and adult film status for movies that are adult films and the genre of movie shared between the hollywood and bollywood datasets. It is ordered by the primary title and limited to 10 results.

```python
%%bigquery
select debut_aslead_role
from kaggle_staging.bollywood_actors AS actor
JOIN kaggle_staging.bollywood_actress AS actress
ON actor.Height_in_cm_ = actress.Height_in_cm_
WHERE debut_aslead_role LIKE 'A%'
ORDER BY debut_aslead_role
LIMIT 10
```

04/15/2021
Returns the run time in minutes for movies where they're not one or ten minutes long and the genre of movie shared between the hollywood and bollywood datasets. It is ordered in ascending order by the runtime and limited to 15 results.

```python
%%bigquery
select runtimeMinutes
from IMDb_staging.title_basics AS hollywood
JOIN kaggle_staging.bollywood AS bollywood
ON hollywood.genres = bollywood.genre
WHERE runtimeMinutes != "1" AND runtimeMinutes != "10"
ORDER BY runtimeMinutes ASC
LIMIT 15
```

04/15/2021
Performs a self join to return the bollywood director and bollywood title for movies that are directed S.N. Patankar or Shree Nath Patankar. It is ordered by the title and limited to 10 results.

```python
%%bigquery
select d1.director, d1.title
from kaggle_staging.bollywood AS d1
JOIN kaggle_staging.bollywood AS d2
ON d1.director = d2.director
WHERE d1.director = "S.N. Patankar" OR d1.director = "Shree Nath Patankar"
ORDER BY title
LIMIT 10
```

04/15/2021
Returns the year and genre for movies whose start year was in 2000 and the start year of the movie is shared between the hollywood and bollywood datasets. It is ordered by the genre and limited to 10 results.

Join was not possible due to incompatible data types INT and STRING. To make the initial construction of the tables, we had to explicitly type some variables as strings that would normally be integers. Therefore, to make the join possible, these changed variables - namely startYear in the hollywood dataset - must be correctly typed as an integer.

```python
%%bigquery
select year, genre
from kaggle_staging.bollywood AS bollywood
JOIN IMDb_staging.title_basics AS hollywood
ON bollywood.year = hollywood.startYear
WHERE hollywood.startYear = 2000
ORDER BY genre
LIMIT 10
```

04/15/2021
Returns the unique title identifier and runtime in minutes for movies where the runtime is not 0 and not null and the titleId of the movie shared between the title_principals and title_basics tables of the hollywood dataset. It is ordered by the runtime in ascending order and limited to 10 results.

```python
%%bigquery
select title.tconst, runtimeMinutes
from IMDb_staging.title_basics AS title
JOIN IMDb_staging.title_principals AS p
ON title.tconst = p.tconst
WHERE runtimeMinutes != "0" AND runtimeMinutes IS NOT NULL
ORDER BY runtimeMinutes ASC
LIMIT 10
```

04/15/2021
Returns the original title and adult status where the movie is not an adult movie and the titleId of the movie is shared between the title_basics and title_ratings tables of the hollywood dataset. It is ordered by the original title and limited to 10 results.

```python
%%bigquery
select isAdult, originalTitle
from IMDb_staging.title_basics AS basics
JOIN IMDb_staging.title_ratings AS ratings
ON basics.tconst = ratings.tconst
WHERE isAdult = "0"
ORDER BY originalTitle
LIMIT 10
```

Returns the actor name and height for bollywood actors that are taller than 100 cm for those that have acted in both hollywood and bollywood It is ordered by the height and limited to 10 results.

```python
%%bigquery
select bollyactor.name, bollyactor.Height_in_cm_
from kaggle_staging.bollywood_actors AS bollyactor
JOIN IMDb_staging.name_basics AS hollyactor
ON bollyactor.name = hollyactor.primaryName
WHERE bollyactor.Height_in_cm_ > 100
ORDER BY bollyactor.Height_in_cm_
LIMIT 10
```

04/15/2021
Returns the actor name and birth year for hollywood actors whose name ends in 'a' and the nconst is shared between the name_basics and title_principals tables of the hollywood dataset. It is ordered by the actor's birth year and limited to 10 results.

```python
%%bigquery
select birthYear, primaryName
from IMDb_staging.name_basics AS basics
JOIN IMDb_staging.title_principals AS p
ON basics.nconst = p.nconst
WHERE primaryName LIKE '%a'
ORDER BY birthYear
LIMIT 10
```

Returns the actor name and height for bollywood actors that are shorter than 170 cm for bollywood actors and actresses are the same height. It is ordered by the height in ascending order and limited to 10 results.

```python
%%bigquery
select actor.name, actor.Height_in_cm_ as height
from kaggle_staging.bollywood_actors AS actor
JOIN kaggle_staging.bollywood_actress AS actress
ON actor.Height_in_cm_ = actress.Height_in_cm_
WHERE actor.Height_in_cm_ < 170
ORDER BY actor.Height_in_cm_ ASC
LIMIT 10
```
